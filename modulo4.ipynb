**Importe as bibliotécas Pandas e Numpy.**

```python
import pandas as pd
import numpy as np
```

**Criar dados fictícios**

```python
dados = {
    'dia': pd.date_range(start='2024-01-01', periods=7),
    'horas_trabalhadas': np.random.randint(4, 9, size=7),
    'bugs_corrigidos': np.random.randint(0, 5, size=7),
    'tarefas_concluidas': np.random.randint(5, 15, size=7)
}
```

**Criar DataFrame com os dados**


```python
df = pd.DataFrame(dados)
df
```

**Calcular métricas**


```python
total_horas_trabalhadas = df['horas_trabalhadas'].sum()
media_diaria_horas_trabalhadas = df['horas_trabalhadas'].mean()
total_bugs_corrigidos = df['bugs_corrigidos'].sum()
media_diaria_bugs_corrigidos = df['bugs_corrigidos'].mean()
total_tarefas_concluidas = df['tarefas_concluidas'].sum()
media_diaria_tarefas_concluidas = df['tarefas_concluidas'].mean()
produtividade_diaria = total_tarefas_concluidas / total_horas_trabalhadas
```

**Criar relatório**


```python
relatorio = pd.DataFrame({
    'Total de Horas Trabalhadas': [total_horas_trabalhadas],
    'Média Diária de Horas Trabalhadas': [media_diaria_horas_trabalhadas],
    'Total de Bugs Corrigidos': [total_bugs_corrigidos],
    'Média Diária de Bugs Corrigidos': [media_diaria_bugs_corrigidos],
    'Total de Tarefas Concluídas': [total_tarefas_concluidas],
    'Média Diária de Tarefas Concluídas': [media_diaria_tarefas_concluidas],
    'Produtividade Diária': [produtividade_diaria]
})
relatorio
```

**Importe a bibliotéca Matplotlib.pyplot**

```python
import matplotlib.pyplot as plt
```

***Como variam as horas trabalhadas por dia?***

```python
df.plot(x='dia', y='horas_trabalhadas', kind='bar')
plt.title('Total de Horas Trabalhadas por Dia')
plt.xlabel('Dia')
plt.ylabel('Horas Trabalhadas')
plt.show()
```

***Existe alguma relação entre as horas trabalhadas e os bugs corrigidos?***

```python
df.plot(x='horas_trabalhadas', y='bugs_corrigidos', kind='bar')
plt.title('Relação entre Horas Trabalhadas e Bugs Corrigidos')
plt.xlabel('Horas Trabalhadas')
plt.ylabel('Bugs Corrigidos')
plt.show()
```
